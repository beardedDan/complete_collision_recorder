```python
import os
import time
import numpy as np
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# Set up Selenium WebDriver
download_dir = os.path.abspath(
    "../data/raw/cad_pdf_files"
)  # Absolute path to the download folder
chrome_options = Options()
chrome_options.add_experimental_option(
    "prefs",
    {
        "download.default_directory": download_dir,  # Set download directory
        "download.prompt_for_download": False,  # Disable prompt for download
        "download.directory_upgrade": True,
        "plugins.always_open_pdf_externally": True,  # Auto-download PDFs instead of opening in Chrome
    },
)

# Initialize the WebDriver (Make sure you have the ChromeDriver or GeckoDriver in your PATH)
driver = webdriver.Chrome(options=chrome_options)

# Assume EVENT_ID is your numpy array of event IDs
EVENT_ID = filtered_traffic_ids

# Base URL for concatenation
base_url = "https://czt313d2a2.execute-api.us-east-1.amazonaws.com/dev/cad/"

# Folder where the PDF files will be saved
output_folder = "../data/raw/cad_pdf_files/"
os.makedirs(
    output_folder, exist_ok=True
)  # Create the folder if it doesn't exist

# Paths to central files for found and not found PDFs
found_pdfs_file = "../data/raw/cad_pdf_files/found_pdfs.txt"
not_found_pdfs_file = "../data/raw/cad_pdf_files/not_found_pdfs.txt"

# Ensure the central files exist
open(found_pdfs_file, "a").close()  # Create if doesn't exist
open(not_found_pdfs_file, "a").close()  # Create if doesn't exist

# Load already processed IDs from not_found_pdfs.txt
with open(not_found_pdfs_file, "r") as f:
    not_found_ids = set(line.strip() for line in f)

with open(found_pdfs_file, "r") as f:
    found_ids = set(line.strip() for line in f)


# Function to download the PDF using Selenium
def download_pdf(event_id):
    pdf_url = f"{base_url}{event_id}.pdf"
    # print(f"Attempting to download: {pdf_url}")
    try:
        # Open the URL in Selenium
        driver.get(pdf_url)

        # Wait for a few seconds to allow the PDF to download
        time.sleep(5)  # Adjust this if needed

        # Check if the file exists in the download directory
        local_filename = os.path.join(output_folder, f"{event_id}.pdf")
        if os.path.isfile(local_filename):
            # Add event_id to found_pdfs.txt
            with open(found_pdfs_file, "a") as f_found:
                f_found.write(f"{event_id}\n")
            # print(f"Downloaded: {local_filename}")
        else:
            # Add event_id to not_found_pdfs.txt if file not found
            with open(not_found_pdfs_file, "a") as f_not_found:
                f_not_found.write(f"{event_id}\n")
            # print(f"File not found for EVENT_ID {event_id}.")
    except Exception as e:
        # Add event_id to not_found_pdfs.txt in case of an error
        with open(not_found_pdfs_file, "a") as f_not_found:
            f_not_found.write(f"{event_id}\n")
        # print(f"Error downloading {event_id}: {e}")


# Total number of IDs
total_ids = len(EVENT_ID)
percent_step = total_ids // 100  # 1% progress
next_percent = percent_step

# Iterate over each EVENT_ID in the numpy array and attempt to download the PDF
for idx, event_id in enumerate(EVENT_ID, 1):
    # Skip if the event_id is already in not_found_pdfs.txt
    if str(event_id) in not_found_ids:
        continue
    if str(event_id) in found_ids:
        continue

    download_pdf(event_id)

    # Print progress every 1%
    if idx >= next_percent:
        print(
            f"{(idx / total_ids) * 100:.2f}% processed ({idx} of {total_ids})"
        )
        next_percent += percent_step

# Close the browser after the process is completed
driver.quit()

print("Process completed.")
```
