```python
import os
import requests

# Path to the file containing the URLs
file_path = "../data/raw/index_files/traffic_event_indexes.txt"

# Local folder to save the downloaded CSV files
output_folder = "../data/raw/index_files/csv_files"

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)


# Function to download a file from a URL with a 30-second timeout
def download_file(file_url, output_folder):
    # Extract the filename from the URL
    local_filename = os.path.join(output_folder, file_url.split("/")[-1])

    try:
        # Request the file with a 30-second timeout
        with requests.get(file_url, stream=True, timeout=30) as r:
            r.raise_for_status()
            # Save the file to the specified location
            with open(local_filename, "wb") as f:
                for chunk in r.iter_content(chunk_size=8192):
                    f.write(chunk)
        print(f"Downloaded: {local_filename}")
    except requests.exceptions.Timeout:
        # Handle timeout without traceback, just print a message
        print(f"Timeout occurred for: {file_url}. Skipping to the next file.")
    except requests.exceptions.RequestException as e:
        # Handle other request errors without traceback
        print(f"Error downloading {file_url}: {e}. Skipping to the next file.")


# Iterate through each line in the file and download the CSV
with open(file_path, "r") as f:
    for line in f:
        csv_url = line.strip()  # Remove any extra whitespace
        if csv_url:  # Ensure the line is not empty
            download_file(csv_url, output_folder)

print("All files processed.")
```
